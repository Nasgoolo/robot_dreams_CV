```python
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.animation as animation

# Set the figure size for plots
plt.rcParams['figure.figsize'] = [8, 6]
```

```python
# Specify the folder containing the frames
folder = 'data/traffic'

# Get the list of frames in the folder
frames = sorted(os.listdir(folder))

# Print the count of frames
print("Frames count:", len(frames))
```

```python
# Specify the index of the frame to use as a tracking template
idx = frames.index('traffic_0.jpg')
```

```python
# Specify the object ID
obj_id = 1

# Specify the coordinates of the bounding box
x1, y1 = 24, 139
x2, y2 = 470, 522

# Calculate the width and height of the bounding box
width = x2 - x1
height = y2 - y1
```

```python
# Load the tracking template
template = cv2.imread(os.path.join(folder, frames[idx]))
template = cv2.cvtColor(template, cv2.COLOR_BGR2RGB)
template = template[y1:y2, x1:x2] / 255

# Display the tracking template
plt.imshow(template)
```

```python
# Create a list to store the processed images
imgs = []

# Specify the number of frames to process
num_frames = 15 

# Set up the tracker
tracker_types = ['MIL', 'KCF',  'CSRT']
tracker_type = tracker_types[1]

# Initialize the tracker based on the selected type
tracker = None
if tracker_type == 'MIL':
    tracker = cv2.TrackerMIL_create()
elif tracker_type == 'KCF':
    tracker = cv2.TrackerKCF_create()
elif tracker_type == "CSRT":
    tracker = cv2.TrackerCSRT_create()

# Initialize the tracker with the bounding box
bbox = (x1, y1, width, height)
im = cv2.imread(os.path.join(folder, frames[idx]))
ok = tracker.init(im, bbox)

# Tracking loop
for ii in range(idx, idx + num_frames):
    # Load the current frame
    img = cv2.imread(os.path.join(folder, frames[ii]))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Update the tracker with the current frame
    ok, bbox = tracker.update(img)
    print("Frame ", ii, " : ", ok, bbox)

    # Show the bounding box on the image
    x1, y1 = int(bbox[0]), int(bbox[1])
    width, height = int(bbox[2]), int(bbox[3])
    imgs.append(cv2.rectangle(img, (x1, y1), (x1 + width, y1 + height), (0, 255, 0), 2))

# Print the count of processed images
print(len(imgs))
```

```python
# Specify the output folder for saving the frames
output_folder = 'data/output_frames'

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)


# Iterate over the frames and print the bounding box on each frame
for i in range(num_frames):
    # Load the current frame
    frame = frames[i]
    img = cv2.imread(os.path.join(folder, frame))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Update the tracker with the current frame
    ok, bbox = tracker.update(img)

    # Show the bounding box on the image
    x1, y1 = int(bbox[0]), int(bbox[1])
    width, height = int(bbox[2]), int(bbox[3])
    img = cv2.rectangle(img, (x1, y1), (x1 + width, y1 + height), (0, 255, 0), 2)

    # Save the modified image with the bounding box
    output_file = os.path.join(output_folder, f'{tracker_type}_frame_{i}.jpg')
    cv2.imwrite(output_file, img)

    print("frame", i, "saved with bounding box.")

# Print the count of processed frames
print(num_frames, "frames processed.")
```

```python
Questions
Do you notice any differences? If so, what are they?
    - Yes, there are. All trackers behave quite differently. The best performance 
    we get is when an object has linear motion without rotation. Some trackers lose 
    the object completely, others can get stuck or switch to another object.

Is one tracker better than another? In what way?
    - All trackers have different speeds and accuracies. And can be used for different cases.
    MIL - works well with partial occlusions, but may have difficulty with large changes in appearance. 
    KCF - works well with different types of objects, but can be sensitive to occlusions. 
    CSRT - works well in complex scenarios with occlusions, scale changes and fast motion.
```

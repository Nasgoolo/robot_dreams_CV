```python
import os
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, Dense, Dropout, Flatten, Input, MaxPooling2D, BatchNormalization
from tensorflow.keras import Model
from tensorflow.keras import layers
from tensorflow.keras.utils import plot_model

from time import time

from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]

from sklearn.utils import shuffle

# Set the seeds for reproducibility
from numpy.random import seed
from tensorflow.random import set_seed
seed_value = 1234578790
seed(seed_value)
set_seed(seed_value)
```

```python
#Define input/output size for data generator
input_size = 64
output_size = 250

# Paths
def get_path(*args):
    return os.path.join('data', *args)

data_path = get_path('images')
```

```python
#Loading data
samples = os.listdir(data_path)
np.random.shuffle(samples)

#80% for train
train_data = samples[:int(len(samples) * 0.8)]
val_test_data = samples[int(len(samples)*0.8):len(samples)]

#10% for validation
validation_data = samples[int(len(samples) * 0.8):int(len(samples) * 0.9)]
#10% for test
test_data = samples[int(len(samples) * 0.9):]

print("Samples count: ", len(samples))
print("Train samples: ", len(train_data))
print("Validation samples: ", len(validation_data))
print("Test samples: ", len(test_data))
```

```python
#Inspecting database
count = 4
for x in range(count):
    img = cv2.cvtColor(cv2.imread(os.path.join(data_path, np.random.choice(samples))), cv2.COLOR_BGR2RGB)
    plt.subplot(1,count,x+1), plt.imshow(img)
```

```python
# Function to prepare data for data generator
def prepare_data(item):
    img = cv2.imread(os.path.join(data_path, item))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    rows, cols, colors = img.shape

    # Compute random crop coordinates
    crop = [np.random.randint(rows - output_size), np.random.randint(cols - output_size)]

    # Crop the image
    img = img[crop[0]:crop[0]+output_size, crop[1]:crop[1]+output_size]

    # Normalize image values
    img = img / 255

    y = img
    x = cv2.resize(img, (input_size, input_size))    

    return x, y
```

```python
#check prepare data function
a,b = prepare_data(train_data[1])

plt.subplot(121), plt.imshow(a)
plt.subplot(122), plt.imshow(b)
```

```python
# Data generator
class Generator(tf.keras.utils.Sequence):    
    def __init__(self, batch_size, input_img_paths):
        self.batch_size = batch_size
        self.input_img_paths = input_img_paths

    def __len__(self):
        return len(self.input_img_paths) // self.batch_size

    def __getitem__(self, idx):
        x_ = []
        y_ = []
        offset = idx*self.batch_size 
        batchPaths = self.input_img_paths[offset:offset+self.batch_size]
        
        for item in batchPaths:
            x, y = prepare_data(item)
            x_.append(x)
            y_.append(y)

        x_ = np.array(x_)
        y_ = np.array(y_)
            
        return x_, y_
    
    def on_epoch_end(self):
        self.input_img_paths = shuffle(self.input_img_paths)
```

```python
def plot_history(history):
    h = history.history
    epochs = range(len(h['loss']))

    plt.subplot(121), plt.plot(epochs, h['loss'], '.-', epochs, h['val_loss'], '.-')
    plt.grid(True), plt.xlabel('epochs'), plt.ylabel('loss')
    plt.legend(['Train', 'Validation'])
    plt.subplot(122), plt.plot(epochs, h['PCNR'], '.-',
                               epochs, h['val_PCNR'], '.-')
    plt.grid(True), plt.xlabel('epochs'), plt.ylabel('PCNR')
    plt.legend(['Train', 'Validation'])
        
    print('Train PCNR     ', h['PCNR'][-1])
    print('Validation PCNR', h['val_PCNR'][-1])
```

```python
#Building model
kernels_count = 128
residual_blocks = 8

def ResidualBlock(inputs, kernels, size):
    x = tf.keras.layers.Conv2D(kernels, size, padding="same", activation="relu")(inputs)
    x = tf.keras.layers.Conv2D(kernels, size, padding="same")(x)
    x = tf.keras.layers.Add()([x, inputs])
    x = tf.keras.layers.Activation('relu')(x)
    return x

def UpscaleBlock(inputs, kernels, size):
    x = tf.keras.layers.Conv2DTranspose(kernels, size, strides=2, padding="same", activation="relu")(inputs)
    return x

def GetModel(kernels_count):
    k = kernels_count

    inputs = tf.keras.layers.Input(shape=(input_size, input_size, 3))
    x = tf.keras.layers.Conv2D(k, 5, padding="same", activation="relu")(inputs)

    for count in range(residual_blocks):
        x = ResidualBlock(x, k, 3)

    x = UpscaleBlock(x, k, 3)

    for count in range(residual_blocks):
        x = ResidualBlock(x, k, 3)

    x = UpscaleBlock(x, k, 3)

    outputs = tf.keras.layers.Conv2D(3, 1, strides=1, padding="same", activation="sigmoid")(x)

    return tf.keras.Model(inputs, outputs)

model = GetModel(kernels_count)
model.summary()
```

```python
#training setup
batch_size = 32
epochs_count = 300

steps_epoch = len(train_data) // batch_size

train_gen = Generator(batch_size, train_data)
validation_gen = Generator(batch_size, validation_data)

print("Epoch steps: ", steps_epoch)
```

```python
# Choice of optimizer, loss function and metrics for model compilation
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001) # Choosing an Adam optimizer with a learning rate of 0.001.
loss_function = tf.keras.losses.MeanSquaredError() # Using the root mean squared error as a loss function
metrics = [tf.keras.metrics.MeanSquaredError(name='MSE'), tf.keras.metrics.RootMeanSquaredError(name='RMSE')] #Defining RMS and RMS as Evaluation Metrics

# Compile the model
model.compile(optimizer=optimizer, loss=loss_function, metrics=metrics)
```

```python
#training
history = model.fit(train_gen, epochs=epochs_count, steps_per_epoch=steps_epoch, validation_data=validation_gen)
```

```python
plot_history(history)
```

```python
#create generator for test
test_gen = Generator(batch_size, test_data)

#evaluate model
pred_test = model.evaluate(test_gen)
```
